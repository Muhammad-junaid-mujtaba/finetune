```python
tokenizer.add_tokens(["!"])  # Ensure it handles exclamation marks correctly
model.resize_token_embeddings(len(tokenizer))
```

```python
def tokenize_function(example):
    user_input_encodings = tokenizer(
        example["user_id"],
        truncation=True,
        padding="max_length",
        max_length=128,
        # return_tensors="pt"  # Remove this line to return lists instead of tensors
    )
    bot_response_encodings = tokenizer(
        example["bot_response"],
        truncation=True,
        padding="max_length",
        max_length=128,
       # return_tensors="pt"  # Remove this line to return lists instead of tensors
    )

    return {
        **user_input_encodings,
        "labels": bot_response_encodings["input_ids"],
    }
dataset = Dataset.from_pandas(df)
tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Convert to PyTorch tensors after tokenization
tokenized_datasets = tokenized_datasets.remove_columns(["user_id", "bot_response"])
tokenized_datasets.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=16,
    lora_dropout=0,
    target_modules=["q_proj", "k_proj", "v_proj", "up_proj", "down_proj", "o_proj", "gate_proj"],
    use_rslora=True,
    use_gradient_checkpointing="unsloth",
    random_state = 32,
    loftq_config = None,
)
print(model.print_trainable_parameters())
```

```python
data_prompt = """Analyze the provided text from a mental health perspective. Identify any indicators of emotional distress, coping mechanisms, or psychological well-being. Highlight any potential concerns or positive aspects related to mental health, and provide a brief explanation for each observation.

### Input:
{}

### Response:
{}"""
EOS_TOKEN = tokenizer.eos_token
def formatting_prompt(examples):
    inputs       = examples["Context"]
    outputs      = examples["Response"]
    texts = []
    for input_, output in zip(inputs, outputs):
        text = data_prompt.format(input_, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
```

```python
from trl import SFTTrainer, AutoModelForCausalLMWithValueHead
from unsloth import is_bfloat16_supported
df['text'] = df.apply(lambda row: data_prompt.format(row['user_id'], row['bot_response']) + EOS_TOKEN, axis=1)
dataset = Dataset.from_pandas(df)

# Now, the 'text' column is available in your dataset.
tokenized_datasets = dataset.map(
    lambda examples: tokenizer(examples['text'], truncation=True, padding="max_length", max_length=128),
    batched=True
)
trainer=SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=tokenized_datasets,
    dataset_text_field="text",
    max_seq_length=64,
    dataset_num_proc=2,
    packing=True,
    args=TrainingArguments(
        learning_rate=3e-4,
        output_dir="./results",
        label_smoothing_factor=0.1,
        lr_scheduler_type="linear",
        per_device_train_batch_size=16,
        gradient_accumulation_steps=8,
        num_train_epochs=10,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=1,
        optim="adamw_8bit",
        weight_decay=0.01,
        warmup_steps=10,
        seed=0,
    ),
)

trainer.train()
```

```python
trainer.save_model("fine_tuned_llama_model")
```

```python
from transformers import BlenderbotForConditionalGeneration, BlenderbotTokenizerFast, pipeline
import torch

# Load model and tokenizer and move to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BlenderbotForConditionalGeneration.from_pretrained("./fine_tuned_model").to(device)
tokenizer = BlenderbotTokenizerFast.from_pretrained("./fine_tuned_model")


def generate_response(user_input):
    # Ensure input tensors are on the same device as the model
    inputs = tokenizer(user_input, return_tensors="pt").to(device)
    reply_ids = model.generate(inputs["input_ids"], max_length=50)
    response = tokenizer.decode(reply_ids[0], skip_special_tokens=True)
    return response

user_input = input("enter somrthing")
print(f"User: {user_input}")
response = generate_response(user_input)
print(f"Bot: {response}")
```
