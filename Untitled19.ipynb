```python
import pandas as pd
from datasets import Dataset


df = pd.read_csv('/content/dialogs.txt', delimiter='\t', on_bad_lines='skip', names=['user_id', 'bot_response'])

df
dataset = Dataset.from_pandas(df)
dataset
```

```python
df
```

```python
def tokenize_function(example):

    user_input_encodings = tokenizer(example["user_id"], truncation=True, padding="max_length", max_length=128)
    bot_response_encodings = tokenizer(example["bot_response"], truncation=True, padding="max_length", max_length=128)


    return {
        **user_input_encodings,
        "labels": bot_response_encodings["input_ids"],
    }

tokenized_datasets = dataset.map(tokenize_function, batched=True)
dataset = Dataset.from_pandas(df)

tokenized_datasets = tokenized_datasets.remove_columns(["user_id", "bot_response"])


tokenized_datasets.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
! pip install peft
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=16,
    lora_dropout=0,
    target_modules=["q_proj", "k_proj", "v_proj", "up_proj", "down_proj", "o_proj", "gate_proj"],
    use_rslora=True,
    use_gradient_checkpointing="unsloth",
    random_state = 32,
    loftq_config = None,
)
print(model.print_trainable_parameters())
```

```python
data_prompt = """Analyze the provided text from a mental health perspective. Identify any indicators of emotional distress, coping mechanisms, or psychological well-being. Highlight any potential concerns or positive aspects related to mental health, and provide a brief explanation for each observation.

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token
def formatting_prompt(examples):
    inputs       = examples["Context"]
    outputs      = examples["Response"]
    texts = []
    for input_, output in zip(inputs, outputs):
        text = data_prompt.format(input_, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
```

```python
from trl import SFTTrainer, AutoModelForCausalLMWithValueHead
from unsloth import is_bfloat16_supported
trainer=SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=tokenized_datasets,
    dataset_text_field="text",
    max_seq_length=128,
    dataset_num_proc=2,
    packing=True,
    args=TrainingArguments(
        learning_rate=3e-4,
        lr_scheduler_type="linear",
        per_device_train_batch_size=16,
        gradient_accumulation_steps=8,
        num_train_epochs=10,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=1,
        optim="adamw_8bit",
        weight_decay=0.01,
        warmup_steps=10,
        output_dir="output",
        seed=0,
    ),
)

trainer.train()
```

```python
trainer.save_model("fine_tuned_llama_model")
```

```python

from transformers import BlenderbotForConditionalGeneration, BlenderbotTokenizerFast

model = BlenderbotForConditionalGeneration.from_pretrained("./fine_tuned_model")
tokenizer = BlenderbotTokenizerFast.from_pretrained("./fine_tuned_model") # Changed to BlenderbotTokenizerFast


def generate_response(user_input):
    inputs = tokenizer(user_input, return_tensors="pt")
    reply_ids = model.generate(inputs["input_ids"], max_length=50)
    response = tokenizer.decode(reply_ids[0], skip_special_tokens=True)
    return response

user_input = input("enter somrthing")
print(f"User: {user_input}")
response = generate_response(user_input)
print(f"Bot: {response}")
```
